```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import requests
from datetime import datetime
import os, time
```

```python
# Directory to save the images
save_folder = 'D:/Loyalist Subjects/Web Scraping/Birds'
if not os.path.exists(save_folder):
    os.makedirs(save_folder)
```

```python
def generate_image_name():
    return datetime.now().strftime("%d-%m-%Y-%H-%M-%S-%f") + ".jpg"
```

```python
def setup_driver():
    # Setup WebDriver (Chrome in this case)
    webdriver_service=Service("C:\driver\chromedriver.exe")
    driver = webdriver.Chrome(service = webdriver_service)
    return driver

def scrape_images(driver, url):
    driver.get(url)
    # This time Waits for the images to load.
    time.sleep(5)  
    image_elements = driver.find_elements(By.TAG_NAME, 'img')
    # Filter images by src attribute
    jpg_png_images = [
        img.get_attribute('src') for img in image_elements 
        if img.get_attribute('src') 
    ]
    print(jpg_png_images)
    return jpg_png_images

def main(base_url, start_page, end_page):
    driver = setup_driver()
    all_images = []
    page = 1
    while True:
        url = base_url.format(page=page)
        images = scrape_images(driver, url)
        download_images(images)
        all_images.extend(images)
        print(f"Scraped {len(images)} images from page {page}.")
        page += 1
    
    print(f"Total images scraped: {len(all_images)}")
    
    driver.quit()

def download_images(jpg_png_images):
    # Loop through each image URL and save the images
    for image_url in jpg_png_images:
        # Get the image content
        response = requests.get(image_url)
        print("response", response)
        if response.status_code == 200:  # Check if the request was successful
            # Extract image name from URL or generate a unique name
            image_name = generate_image_name()
            print(image_name)
            # Create the full path to save the image
            save_path = os.path.join(save_folder, image_name)

            # Write the image content to a file in the specified directory
            with open(save_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded {image_name} to {save_path}")
        else:
            print(f"Failed to retrieve image at {image_url}")
    
    
if __name__ == "__main__":
    BASE_URL = "https://www.istockphoto.com/search/2/image-film?istockcollection=&mood=&page={page}&phrase=myna"
    START_PAGE = 1
    END_PAGE = 5  # Adjust the end page as per requirement
    main(BASE_URL, START_PAGE, END_PAGE)
```
